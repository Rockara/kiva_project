## Step 4: Bar Charts

Create a bar plot using Seaborn to visualize the average size of Kiva loans given to projects, by country.

We've set up the figure you'll use to plot your bar plot on. The `f` variable gives us access to the figure and `ax` gives us access to the axes.

Use `sns.barplot()` with the following arguments:

- `data` set to `df`
- `x` set to `country`
- `y` set to `loan_amount`

````python
# Creates the figure, note: you're only using this syntax so that you can modify the y-axis ticks later
plt.figure(figsize=(30,20))
ax = plt.subplot(2,1,1)
ax.set_title('Each country\'s loans')
ax = sns.barplot(data=df, x='country', y='loan_amount', palette='pastel')

ax2 = plt.subplot(2,1,2)
ax2.set_title('Each country\'s loans by gender')
ax2 = sns.barplot(data=df, x='country', y='loan_amount', hue='gender', palette='dark')
````

## Step 8: Box Plot by Activity

Instead of visualizing the loan amount by *country*, use `sns.boxplot()` to plot the loan amount by *activity*.
<br>
<br>
<details>
<summary>Hint</summary>
<br>
You can use the same code as the box plot above, but the `x` parameter should be set to `"activity"`.
<br>
</details>
 
**Optional:** Set a different plot style and color palette to best visualize this data.

````python
print(df.head())

plt.figure(figsize=(16, 20))

ax = plt.subplot(2,1,1)
ax = sns.boxplot(data=df, x='activity', y='loan_amount')

df_female = df[df.gender == 'female']
fe_general = df_female[df_female.activity == 'General Store']
fe_food = df_female[df_female.activity == 'Food Production/Sales']
fe_farm = df_female[df_female.activity == 'Farming']
general_count = len(fe_general)
food_count = len(fe_food)
farming_count = len(fe_farm)
print("Female reasons for loans: \ngeneral count: ", general_count, "\nfood count: ", food_count, "\nFarming count: ", farming_count)
# well well, this perhaps give some clarity on my formar questions
# looking at this print we see general store/sales is the absolutly most common reason to take loans for females
# this can obviusly mean far more things than just clothing. but clothing seems to be the far most likely
# since food is a separate category, as is farming.

ax2 = plt.subplot(2,1,2)
ax2 = sns.boxplot(data=df_female, x='activity', y='loan_amount')

# the average loan for females in for the general store seems to be 350:- requests. About the price of 
# where I would make a guess the average expencive clothing piece would be.

# however, lets look at the male counts as well.

df_male = df[df.gender == 'male']
male_general = len(df_male[df_male.activity == 'General Store'])
male_food = len(df_male[df_male.activity == 'Food Production/Sales'])
male_farm = len(df_male[df_male.activity == 'Farming'])
print('\nMale reasons for loans: \nGeneral_count: ', male_general, '\nfood count: ', male_food, '\nfarming count: ', male_farm)

# here we can also clearly see that men take way fewer loans than females
# but the loans men take are far higer. this does give a lot of credibility to my hypotesis
````
